```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from decomposition.game import Game
```

Let's consider randomly generated games by choosing each entry of the payoff matrix from a uniform distribution. First we define the general structure of the game:

```python
n_actions = [3, 3] # 2 agents with 3 actions each
game = Game(n_actions, save_load=True, path="../")
```

Now generate 100_000 random games and check the distribution of the potentialness:

```python
result = []
for _ in tqdm(range(100_000)):
    payoff_vector = np.random.rand(game.n_agents * np.prod(game.n_actions))
    game.compute_decomposition(payoff_vector)
    result.append(game.metric)
```

```python
plt.hist(result, density=True, bins=50)
plt.ylabel("Relative Frequency")
plt.xlabel("Potentialness")
plt.title(f"Potentialness - {game}")
plt.show()
```
